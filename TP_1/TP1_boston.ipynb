```python
# Fonction pour calculer les valeurs manquantes par colonne
def missing_values(df):
        # Total des valeurs manquantes
        mis_val = df.isnull().sum()

        # Pourcentage des valeurs manquantes
        mis_val_percent = 100 * mis_val / len(df)

        # Faites un tableau avec les résultats
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

        # Renomer les colonnes
        mis_val_table = mis_val_table.rename(columns = {0 : 'Valeurs manquantes', 1 : '% du Total des valeurs'})

        # Trier le tableau par pourcentage de décroissant manquant
        mis_val_table = mis_val_table[mis_val_table.iloc[:,1] != 0].sort_values('% du Total des valeurs', ascending=False).round(1)

        # Imprimer des informations récapitulatives
        print ("Votre dataframe sélectionné a " + str(df.shape[1]) + " colonnes.")
        print ("Il y a " + str(mis_val_table.shape[0]) +" colonnes qui ont des valeurs manquantes.")

       # Renvoie le dataframe avec les informations manquantes
        return mis_val_table
```

#### Explication pour la cellule ci-dessous

Ce code extrait les colonnes numériques d'un DataFrame df, calcule la matrice de corrélation entre ces colonnes, et identifie les paires de colonnes qui sont fortement corrélées (avec une corrélation absolue supérieure à un seuil fixé, ici 0.8). Il stocke ensuite ces paires et leurs coefficients de corrélation dans un dictionnaire high_corr_pairs. Cette approche est utile pour détecter les relations linéaires fortes entre les variables, ce qui peut être important pour éviter la multicollinéarité dans les modèles de régression ou pour simplifier un ensemble de données en éliminant des variables redondantes.

```python
# Get numerics data
dfNumeric = data.select_dtypes(include=[np.number])
# Calculate the correlation matrix
correlation_matrix = dfNumeric.corr()

# Find pairs of highly correlated columns
high_corr_pairs = {}
threshold = 0.8  # Set a threshold for strong correlation

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            col_name1 = correlation_matrix.columns[i]
            col_name2 = correlation_matrix.columns[j]
            high_corr_pairs[(col_name1, col_name2)] = correlation_matrix.iloc[i, j]
```

## Application des stratégies de traitement des valeurs manquantes

#### Explication pour la cellule ci-dessous (Variables textuelles)
Variables textuelles remplacées par "Information non disponible" : Cela maintient la clarté des données en indiquant explicitement l'absence d'information, plutôt que de laisser un vide qui pourrait être interprété de manière erronée.

```python
# 1. Variables textuelles: Remplacer par "Information non disponible"
text_vars_with_missing = ['access', 'transit', 'interaction', 'neighborhood_overview', 'house_rules', 'space']
data[text_vars_with_missing] = data[text_vars_with_missing].fillna("Information non disponible")
```

#### Explication pour la cellule ci-dessous (Variables numériques)
on utilise la médiane pour remplir les valeurs manquantes dans les données numériques pour les raisons suivantes :

Robustesse aux valeurs extrêmes : La médiane est moins affectée par les valeurs aberrantes que la moyenne.
Adaptée aux distributions non normales : Elle offre une meilleure mesure de tendance centrale pour les données qui ne sont pas distribuées normalement.
Convient aux scores catégoriels ordinaux : Pour des données comme les scores de review, la médiane peut mieux refléter l'évaluation typique.
Simplicité et interprétabilité : C'est une méthode facile à comprendre et à appliquer.
Préservation de la distribution des données : Utiliser la médiane aide à maintenir la distribution originale des données.

```python
# 2. Variables numériques (Scores de review): Utiliser la médiane
review_score_vars = ['review_scores_checkin', 'review_scores_location', 'review_scores_value',
                     'review_scores_accuracy', 'review_scores_cleanliness',
                     'review_scores_communication', 'review_scores_rating']
for var in review_score_vars:
    data[var] = data[var].fillna(data[var].median())

# 3. Autres variables numériques: Utiliser la médiane
other_numerical_vars_with_missing = ['beds', 'bedrooms', 'bathrooms']
for var in other_numerical_vars_with_missing:
    data[var] = data[var].fillna(data[var].median())
```

#### Explication pour la cellule ci-dessous (Variables de review spécifiques)
reviews_per_month rempli avec 0 : Assigne une valeur neutre aux listings sans reviews, permettant une analyse cohérente sans supposer de l'activité là où il n'y en a pas.

```python
# 4. Variables de review spécifiques (Traitement spécial pour `reviews_per_month`)
data['reviews_per_month'] = data['reviews_per_month'].fillna(0)
```

#### Explication pour la cellule ci-dessous (Variables catégorielles/booléennes)
Variables catégorielles/booléennes remplacées par le mode : Utilise la valeur la plus fréquente pour maintenir la cohérence des caractéristiques générales du dataset, en supposant que la valeur la plus commune est la plus probable en cas d'information manquante.

```python
# 5. Variables catégorielles/booléennes: Remplacer par le mode
categorical_bool_vars_with_missing = ['host_has_profile_pic', 'host_identity_verified', 'host_is_superhost', 'host_response_time']
for var in categorical_bool_vars_with_missing:
    data[var] = data[var].fillna(data[var].mode()[0])
```

#### Explication pour la cellule ci-dessous (security_deposit)
security_deposit, cleaning_fee remplacés par 0 : Traite l'absence de frais comme l'indication qu'aucun frais n'est requis, aligné sur une interprétation conservatrice des coûts.

```python
# 6. `security_deposit`, `cleaning_fee`: Remplacer les valeurs manquantes par zéro
data['security_deposit'] = data['security_deposit'].fillna(0)
data['cleaning_fee'] = data['cleaning_fee'].fillna(0)
```

#### Explication pour la cellule ci-dessous (description et host_location)
Traitement spécial pour description et host_location : Assure que toutes les listings ont des informations de base, même standardisées, pour soutenir des analyses textuelles ou géographiques minimales.

```python
# 7. `description`, `host_location`: Traitement spécial pour la description et la localisation de l'hôte
data['description'] = data['description'].fillna("Information non disponible")
data['host_location'] = data['host_location'].fillna(data['host_location'].mode()[0])
```

```python
# Remplacer 'first_review' et 'last_review' par "non disponible"
data['first_review'].fillna("non disponible", inplace=True)
data['last_review'].fillna("non disponible", inplace=True)
```

#### Explication pour la cellule ci-dessous (host_response_rate)
Conversion de 'host_response_rate' en numérique et remplacement des manquants par la médiane : Fournit une estimation raisonnable de la tendance centrale pour cette variable, permettant une analyse quantitative tout en minimisant l'impact des extrêmes.

Suppression des lignes où host_since est manquant : Élimine les listings avec des informations critiques manquantes sur l'hôte, présumant que l'ancienneté de l'hôte est essentielle pour l'analyse.

```python
# Étape 1: Conversion de 'host_response_rate' en valeurs numériques
data['host_response_rate'] = data['host_response_rate'].str.replace('%', '').astype(float)

# Étape 2: Calcul de la médiane pour 'host_response_rate'
host_response_rate_median = data['host_response_rate'].median()

# Étape 3: Remplacement des valeurs manquantes par la médiane
data['host_response_rate'].fillna(host_response_rate_median, inplace=True)

# Supposer que la suppression des lignes où 'host_since' est manquant est acceptable pour votre analyse :
data.dropna(subset=['host_since'], inplace=True)
```

#### Explication pour la cellule ci-dessous (Conversion des dates et Conversion des variables monétaires)

#### Explication pour la cellule ci-dessous (Encodage des variables)
Encodage des variables binaires en 0/1 : Simplifie l'analyse en convertissant les indicateurs textuels en valeurs numériques binaires, facilitant les comparaisons et les calculs statistiques.

```python
# Encoder les variables binaires 't'/'f' en 0/1
binary_columns = ['host_has_profile_pic', 'host_identity_verified', 'host_is_superhost', 'instant_bookable']

# Boucle sur les colonnes binaires pour les encoder
for col in binary_columns:
    data[col] = data[col].map({'f': 0, 't': 1}).astype(int)
```
